# 3) Data Cleaning

### 3.1) Cleaning Duration

```python
df
```

```python
for i in df['duration'][:10]:
  print(i.split(':'))
```

```python
for i in df['duration'][:10]:
  print(i, int(i.split(':')[0])*60 + int(i.split(':')[1]))
```

```python
duration = []

for i in df['duration']:

  mins = int(i.split(':')[0])
  secs = int(i.split(':')[1])
  duration.append(mins*60 + secs)

df['duration_in_sec'] = duration
df
```

```python
df['duration_in_sec'].sort_values(ascending=False)
```

```python
duration = []

for i in df['duration']:
  if len(i.split(":")) == 3:
    print(i,i.split(':')[0])
```

```python
duration = []

for i in df['duration']:
  if len(i.split(':')) == 2:
    mins = int(i.split(':')[0])
    secs = int(i.split(':')[1])
    duration.append(mins*60 + secs)

  elif len(i.split(':')) == 3:
    hours = int(i.split(':')[0])
    mins = int(i.split(':')[1])
    secs = int(i.split(':')[2])
    duration.append(hours*60*60 + mins*60 + secs)

df['duration_in_sec'] = duration
df
```

```python
df['duration_in_sec'].sort_values(ascending=False)
```
