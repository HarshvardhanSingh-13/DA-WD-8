## Dataset Description

This dataset contains information about various songs, primarily in the Urdu language, along with details about the singer(s), song duration, and a link to the song.

### Columns:

1. **name**: The title of the song.
2. **singer**: The singer(s) associated with the song. In some cases, multiple singers are listed, separated by a pipe (`|`).
3. **singer_id**: A unique identifier for the singer(s), which can be used to link to further information about the artists.
4. **duration**: The length of the song, represented in the format `HH:MM:SS`.
5. **link**: A unique URL identifier for the song that can be used for accessing the song on a platform.
6. **language**: The language in which the song is sung.
### Example:
- The song titled "Dil - E - Nadan Tujhe" is sung by Chitra Singh and Jagjit Singh, with a duration of 5 minutes and 0 seconds, and the song is available in the Urdu language.

This dataset can be useful for music analysis, song recommendation systems, or building databases of music data based on different parameters like artist, language, and song duration.

```python
duration = []

for i in df['duration']:

  mins = int(i.split(':')[0])
  secs = int(i.split(':')[1])
  duration.append(mins*60 + secs)

df['duration_in_sec'] = duration
df
```

```python
df['duration_in_sec'].sort_values(ascending=False)
```

```python
duration = []

for i in df['duration']:
  if len(i.split(":")) == 3:
    print(i,i.split(':')[0])
```

```python
duration = []

for i in df['duration']:
  if len(i.split(':')) == 2:
    mins = int(i.split(':')[0])
    secs = int(i.split(':')[1])
    duration.append(mins*60 + secs)

  elif len(i.split(':')) == 3:
    hours = int(i.split(':')[0])
    mins = int(i.split(':')[1])
    secs = int(i.split(':')[2])
    duration.append(hours*60*60 + mins*60 + secs)

df['duration_in_sec'] = duration
df
```

```python
import matplotlib.pyplot as plt

# Get the top 10 singers and their song counts
top_singers = singers.value_counts().sort_values(ascending=False).head(10)

# Create the bar plot
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(top_singers.index, top_singers.values, color='skyblue', edgecolor='black')

# Customize the plot
plt.title('Top 10 Singers with Most Songs', fontsize=16)
plt.xlabel('Singer', fontsize=12)
plt.ylabel('Number of Songs', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid to the y-axis

# Add data labels to the bars
for i, v in enumerate(top_singers.values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontsize=10)

# Add a tight layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# Get the top 10 singers and their song counts
top_singers = singers.value_counts().sort_values(ascending=False).head(10)

# Normalize the song counts to the range [0, 1] for color mapping
normalized_counts = (top_singers.values - top_singers.values.min()) / (top_singers.values.max() - top_singers.values.min())

# Create the bar plot with shaded color
plt.figure(figsize=(12, 6))
plt.bar(top_singers.index, top_singers.values, color=cm.Reds(normalized_counts), edgecolor='black')  # Use Reds colormap

# Customize the plot
plt.title('Top 10 Singers with Most Songs', fontsize=16)
plt.xlabel('Singer', fontsize=12)
plt.ylabel('Number of Songs', fontsize=12)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels for better readability
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid to the y-axis

# Add data labels to the bars
for i, v in enumerate(top_singers.values):
    plt.text(i, v + 5, str(v), ha='center', va='bottom', fontsize=10)

# Add a tight layout to prevent overlapping elements
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Assuming 'singers' is a pandas Series containing singer names and their song counts
# Get the top 10 singers and their song counts
top_singers = singers.value_counts().sort_values(ascending=False).head(10)

# Create a DataFrame for Plotly
df_top_singers = pd.DataFrame({'Singer': top_singers.index, 'Song Count': top_singers.values})

# Create the bar plot using Plotly Express
fig = px.bar(df_top_singers, x='Singer', y='Song Count',
             title='Top 10 Singers with Most Songs',
             labels={'Singer': 'Singer', 'Song Count': 'Number of Songs'},
             color='Song Count',  # Color bars based on song count
             color_continuous_scale='Blues',  # Use Reds color scale
             text='Song Count')  # Display song count as text on bars

# Customize the layout
fig.update_layout(
    xaxis_tickangle=-45,  # Rotate x-axis labels
    xaxis_title_font={'size': 12},
    yaxis_title_font={'size': 12},
    title_font={'size': 16},
    yaxis_gridcolor='lightgray',  # Add grid to y-axis
    plot_bgcolor='white'  # Set plot background color
)

# Display the plot
fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Assuming 'singers' is a pandas Series containing singer names and their song counts
# Get the top 10 singers and their song counts
top_singers = singers.value_counts().sort_values(ascending=False).head(10)

# Create a DataFrame for Plotly
df_top_singers = pd.DataFrame({'Singer': top_singers.index, 'Song Count': top_singers.values})

# Create the donut chart using Plotly Express
fig = px.pie(df_top_singers, values='Song Count', names='Singer',
             title='Top 10 Singers with Most Songs',
             hole=0.6)  # Set hole size for donut chart

# Customize the layout
fig.update_traces(textposition='inside', textinfo='percent+label')  # Display percentage and label inside the donut slices
fig.update_layout(title_font={'size': 16}, showlegend=True)  # Adjust title font size and show legend

# Display the chart
fig.show()
```
